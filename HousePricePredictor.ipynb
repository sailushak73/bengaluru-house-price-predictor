```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('Bengaluru_House_Data.csv')
data.head(5)
```

```python
data.isna().sum()
```

```python
print(data.columns)
```

```python
data.drop(columns=['area_type','availability','society'], inplace=True)
```

```python
data.head()
```

```python
data.dropna(inplace=True)
```

```python
data.location.value_counts()
```

```python
data['location']=data['location'].apply(lambda x: x.strip())
```

```python
data.location.value_counts()
```

```python
location_stats=data.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
location_less_than_10_entries=location_stats[location_stats <= 10]
location_less_than_10_entries
```

```python
data['location']=data['location'].apply(lambda x: 'other' if x in  location_less_than_10_entries else x)
```

```python
data.location.value_counts()
```

```python
data['size'].unique()
```

```python
data['bedrooms']=data['size'].apply(lambda x:int(x.split(' ')[0]))
```

```python
data
```

```python
data.total_sqft.unique()
```

```python
def clean(sqft):
    values=sqft.split('-')
    if len(values) == 2:
        return (float(values[0])+float(values[1]))/2
    else:
        try:
            return float(sqft)
        except:
            return None
    
```

```python
clean('1134-1384')
```

```python
data['total_sqft']=data['total_sqft'].apply(clean)
```

```python
data.total_sqft.unique()
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
```

```python
data.describe()
```

```python
 data['sqft_per_bed']=data['total_sqft']/data['bedrooms']
```

```python
data.describe()
```

```python
data2=data[data['sqft_per_bed']>=300]
```

```python
data2
```

```python
data2['price_per_sqft']=data2['price']*100000/data2['total_sqft']
```

```python
data2
```

```python
data2.price_per_sqft.describe()
```

```python
data3=data2[data2['price_per_sqft'] >= 2000]
```

```python
data3
```

```python
data3.drop(columns=['size','sqft_per_bed','price_per_sqft'], axis=1, inplace=True)
```

```python
data3
```

```python
%pip install scikit-learn
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.compose import make_column_transformer
```

```python
col_trans=make_column_transformer((OneHotEncoder(sparse_output=False),['location']),remainder='passthrough') 
```

```python
lr=LinearRegression()
```

```python
Scaler=StandardScaler()
```

```python
model=make_pipeline(col_trans,Scaler,lr)
```

```python
data_in=data3.drop(columns=['price'])
data_out=data3['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(data_in,data_out,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
input=pd.DataFrame([['Electronic City Phase II',2000.0,3.0,3.0,39.07,3]],columns=['location','total_sqft','bath','balcony','price','bedrooms'])
```

```python
model.predict(input)
```
